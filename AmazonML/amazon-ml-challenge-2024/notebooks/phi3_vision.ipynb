```python
from PIL import Image
import requests
from transformers import AutoModelForCausalLM
from transformers import AutoProcessor
import torch

model_id = "microsoft/Phi-3-vision-128k-instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="cuda",
    trust_remote_code=True,
    torch_dtype=torch.float16,
    _attn_implementation="eager",
)

processor = AutoProcessor.from_pretrained(model_id, trust_remote_code=True)
```

```python
from PIL import Image
import requests
import torch

# Define the messages used in the prompt
valid_vol = [
    "gram",
    "kilogram",
    "microgram",
    "milligram",
    "ounce",
    "pound",
    "ton",
    "gm",
    "Kg",
    "mg",
    "oz",
]

# Define the messages used in the prompt
messages = [
    {
        "role": "user",
        "content": f"<|image_1|>\nWhat is the maximum weight recommendation mentioned in the image?",
    },
    {
        "role": "assistant",
        "content": f"You are an expert at extracting attributes from the images, and respond using only the following units of maximum weight recommendation: {', '.join(valid_vol)} if they are mentioned in the image.",
    },
]


# Define generation arguments
generation_args = {
    "max_new_tokens": 30,
    "temperature": 0.0,
    "do_sample": False,
}
prompt = processor.tokenizer.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
# Loop through each image URL
```

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/folder_name/maximum_weight_recommendation.csv")
dfw = df
dfw["prediction"] = ""
dfw
```

```python
cnt = 0
for index, row in dfw.iterrows():

    # Load the image
    url = row["image_link"]
    image = Image.open(requests.get(url, stream=True).raw)
    inputs = processor(prompt, [image], return_tensors="pt").to("cuda:0")
    generate_ids = model.generate(
        **inputs, eos_token_id=processor.tokenizer.eos_token_id, **generation_args
    )
    generate_ids = generate_ids[:, inputs["input_ids"].shape[1] :]
    response = processor.batch_decode(
        generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )[0]
    dfw.at[index, "prediction"] = response
    print(f"Response for image at {url}:\n{response}\n")
    file_name = "maximum_weight_recommendation"
    dfw.to_csv(f"/kaggle/working/{file_name}_prediction.csv", index=False)
```
