```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = pd.read_csv('landslide_data_original.csv')
x = (f.groupby('stationid').get_group('t12')['humidity']).tolist()

y = np.arange(min(x),max(x),5)

plt.hist(x)
plt.xticks(y)
plt.grid(visible=True)
plt.xlabel("Humidity intervals -->")
plt.ylabel("Frequency -->")
plt.title("Humidity Histogram for stationidt12")
plt.show()
```
