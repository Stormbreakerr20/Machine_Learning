```python
import numpy as np
x_data = np.array([1.0,2.0])
y_data = np.array([300.0,500.0])


import matplotlib.pyplot as plt

plt.scatter(x_data,y_data, marker='x',c='r')
plt.ylabel("Price")
plt.show()

i = 0
print(f"x^(i), y^(i) = {x_data[i]}, {y_data[i]}")
```

```python
# model
def compute_model_output(x,w,b):
    m = x.shape[0]
    fx = np.zeros(m)
    for i in range(m): fx[i] = x[i]*w + b
    return fx

w = 200   # for w = 100 it does not fit in our curve # only val 200 fits our line into values
b = 100

op_function = compute_model_output(x_data,w,b)

print(op_function)

plt.plot(x_data,op_function,c='b',label = 'Our prediction')
plt.scatter(x_data,y_data, marker = 'x', c='r',label = 'Actual values')
plt.title("Housing Prices")
plt.legend()

plt.show()

        
```

```python
x_req = 1.2
print(f"According to our plot and prediction price for 1.2 sqm = {w*x_req + b} approximatley")
```

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([300,400,550,600,620,750,770])

plt.scatter(x,y,marker='x',c='r')
plt.show()
```

```python
def opf(x,w,b):
    fx = np.zeros(x.shape[0])
    for i in range(x.shape[0]): fx[i] = x[i]*w + b
    return fx

plt.plot(x,opf(x,100,150),c='b')
plt.scatter(x,y,marker='x',c='r')
plt.show()
```
