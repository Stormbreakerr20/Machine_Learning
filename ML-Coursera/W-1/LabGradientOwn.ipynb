```python
import numpy as np
x = np.array([1.0,2.0])
y = np.array([300.0,500.0])

b = 100
w_i = 0

# learning rate
alpha = 1.0e-2
```

```python
def cost_func(x,y,w,b):
    m = x.shape[0]
    sq_error = sum( [(w*x[i] + b - y[i])**2 for i in range(m)] )
    
    return sq_error/(2*m)
```

```python
def compute_gradient(x,y,w,b):
    m = x.shape[0]
    df_w = (1/m)*sum( [(w*x[i] + b - y[i])*x[i] for i in range(m) ] )
    df_b = (1/m)*sum( [(w*x[i] + b - y[i]) for i in range(m) ] )

    return df_w,df_b
```

```python
def gradient_descent(x,y,w,b,aplha):
    
    trials = 1000000

    for i in range(trials):
        df_w,df_b = compute_gradient(x,y,w,b)
        temp_w = w - aplha*df_w
        temp_b = b - alpha*df_b

        w,b = temp_w,temp_b

    return w,b
```

```python
w_final, b_final = gradient_descent(x,y,0,b,alpha)
print(f"Most suited w for linear regression of given trianing set = {w_final}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,marker='x',c='r')
plt.plot(x,w_final*x+b,c='b')
plt.show()
```

```python
cost_func_list = [cost_func(x,y,k,b) for k in range(400)]
w_list = [i for i in range(400)]

plt.plot(w_list,cost_func_list,c='b')
plt.show()
```
