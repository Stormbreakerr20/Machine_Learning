# Handling Categorical And Numerical If Occur in same Col

```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

## Numbers

```python
df['numerical_number'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df['categorical_number'] = np.where(df['numerical_number'].isnull(),df['number'],np.nan)
df.head()
```

## Cabin

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')  # Get Numerical Part from String
df['cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python
df['cabin_cat'].value_counts()
```

## Ticket

```python

# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,df['ticket_cat'])
```
