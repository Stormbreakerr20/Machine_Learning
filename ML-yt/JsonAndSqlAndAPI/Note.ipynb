# Web  Scrapping

```python
# from : https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav
import pandas as pd
import requests as req
from bs4 import BeautifulSoup # lib used for Web Scrapping
import html

req.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav').text
```

```python
#Normal hit is considered as bot hit So now make changes
# add headers
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

webpage = req.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html.parser') #to make parsing easy

print(soup.prettify())
```

```python
# Inspect in browser : get Structure : get div of elements
soup.find_all('h2')
```
