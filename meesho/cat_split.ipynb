```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
for x in df['Category'].unique():
    df1 = df[df['Category'] == x]
    x = x.replace(' ', '_')
    df1.to_csv(f'{x}.csv', index=False)
```

```python
import pandas as pd
df = pd.read_csv('Men_Tshirts.csv')

df = df.rename(columns={
        "attr_1": "color",
    "attr_2": "neck",
    "attr_3": "pattern",
    "attr_4": "print_or_pattern_type",
    "attr_5": "sleeve_length",
})

# Save the updated CSV
df.to_csv("Men_Tshirts.csv", index=False)

df.head()
```
