```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
df = df.rename(columns={
    "attr_1": "color",
    "attr_2": "fit_shape",
    "attr_3": "length",
    "attr_4": "neck_collar",
    "attr_5": "occasion",
    "attr_6": "pattern",
    "attr_7": "print_or_pattern_type",
    "attr_8": "sleeve_length",
    "attr_9": "sleeve_styling",
    "attr_10": "surface_styling"
})

# Save the updated CSV
df.to_csv("train_updated.csv", index=False)
```

```python
df.head()
```

```python
category_summary = {}

for category, group in df.groupby("Category"):
    category_summary[category] = {}
    for col in group.columns[3:]:  # Exclude 'id', 'Category', and 'len'
        unique_values = group[col].dropna().unique()
        category_summary[category][col] = {
            "unique_count": len(unique_values),
            "unique_values": unique_values.tolist()
        }

# Display the summary
for category, attributes in category_summary.items():
    print(f"Category: {category}")
    for attr, info in attributes.items():
        print(f"  {attr}:")
        print(f"    Unique Count: {info['unique_count']}")
        print(f"    Unique Values: {info['unique_values']}")
```
